## Data cleaning for project

```python
#import necessary modules
import pandas as pd
import math
import matplotlib.pyplot as plt
#in addition, we have to install the xlrd and the Seaborn modules in the working environment (done in anaconda)
```

```python
#import dataframe from excel extraction of Sharepoint
Allfiles = pd.read_excel ('DocumentData_ALL.xlsx')
#we have a quick look at it
Allfiles
```

```python
#We get the last 3 years of documents
from datetime import datetime
Allfiles['Created date'] = pd.to_datetime(Allfiles['Created']).dt.year.copy()
Allfiles['Created date']
```

```python
Recent = Allfiles[Allfiles['Created'] > 2017].copy()
Recent
```

```python
#we see the documents by service
#ax = Recent.Service.value_counts().plot(kind="bar",fill=True,label='Service', title='Documents by service')
ax = Recent.Service.value_counts().plot(kind="pie",label='Service', title='Documents by service')
ax.set_xlabel("Service acronym")
ax.set_ylabel("number of documents")
plt.legend()
plt.show()
```

```python
# we proceed with the cleaning
Recent['Name'] = Recent['Name'].str.replace('_',' ').copy()
Recent['Name'] = Recent['Name'].str.replace('/',' ').copy()
Recent['Name'] = Recent['Name'].str.replace('-',' ').copy()
Recent['Name'] = Recent['Name'].str.replace('(','').copy()
Recent['Name'] = Recent['Name'].str.replace(')','').copy()
Recent['Name'] = Recent['Name'].str.replace('.',' ').copy()
#if we wish to remove the filetype from the title, we apply the next line
Recent['Name'] = Recent['Name'].str[:-3].copy()
#and we print the result to see if it worked
Recent['Name']
```

```python
#Now we change the different extensions of the same kind of document, starting by the textual ones
pattern = '|'.join(['dotx', 'docx','docm', 'doc', 'dot','txt', 'css', 'prj'])
Recent['Name'] = Recent['Name'].str.replace(pattern, 'doc').copy()
#and we print the result to see if it worked  
Recent['Name']
```

```python
pattern = '|'.join(['pptx', 'pptm','ppt'])
Recent['Name'] = Recent['Name'].str.replace(pattern, 'ppt').copy()
#and we print the result to see if it worked  
Recent['Name']
```

```python
pattern = '|'.join(['xlsx', 'xlsm','xlsb', 'xls'])
Recent['Name'] = Recent['Name'].str.replace(pattern, 'xls').copy()
#and we print the result to see if it worked  
Recent['Name']
```

```python
pattern = '|'.join(['dotx', 'docx','docm', 'doc', 'dot','txt', 'css', 'prj'])
Recent['Name'] = Recent['Name'].str.replace(pattern, 'doc').copy()
#and we print the result to see if it worked  
Recent['Name']
```

```python
pattern = '|'.join(['pptx', 'pptm','ppt'])
Recent['File_type'] = Recent['File_type'].str.replace(pattern, 'ppt').copy()
#and we print the result to see if it worked  
Recent['File_type']
```

```python
pattern = '|'.join(['xlsx', 'xlsm','xlsb', 'xls'])
Recent['File_type'] = Recent['File_type'].str.replace(pattern, 'xls').copy()
#and we print the result to see if it worked  
Recent['File_type']
```

```python
#we create our final dataframe with the data that we need,
df = Recent[['Name', 'File_type',  'Service']].copy()
df.info()
df.head(10)
```

```python
df.to_csv('CleanTitle.csv')
```
